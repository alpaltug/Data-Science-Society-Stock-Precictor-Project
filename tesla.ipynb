```
df_tesla = df_tesla[df_tesla['lang'] == 'en']
df_tesla = df_tesla[df_tesla['likeCount'] > 2]
```

```
file_object = open('english.txt', 'a')
file_object.write("@")

nltk.download('stopwords')
stop_words = list([set(stopwords.words('english'))])
stop_words.extend(("#", "@", '\'', ':', '.', ',', '!'))

filtered = []
for i in df_tesla['content']:
    i = word_tokenize(i)
    i = [w for w in i if not w.lower() in stop_words]
    for x in i:
        if x in stop_words:
            i.remove(x)
        if "https" or "http" or "//" in x:
            i.remove(x)
    filtered.append(i)

df_tesla["filtered"] = filtered
df_tesla
```

```
list_tesla = []
for i in df_tesla['filtered']:
    blob = TextBlob(str(i))
    list_tesla.append(blob.sentiment.polarity)
           

df_tesla["score"] = list_tesla
df_tesla = df_tesla[['score', 'date', 'filtered', 'replyCount', 'retweetCount', 'likeCount', 'quoteCount']] 
df_tesla = df_tesla[df_tesla['score'] != 0]
# Extracting only date from date column
df_tesla["dateonly"] = pd.to_datetime(df_tesla["date"]).dt.date
df_tesla
```

```
df_datetesla = df_tesla.groupby('dateonly').mean()
```

```
#Stock price data(per day)
from pandas_datareader import data

df_stocktesla = data.get_data_yahoo("TSLA",start="2022-04-01",end="2022-04-23")
df_stocktesla = df_stocktesla.rename_axis('date').reset_index()
df_stocktesla["dateonly"] = pd.to_datetime(df_stocktesla["date"]).dt.date
df_stocktesla
```

```
#Conbine 2 dataframe
df_tesla2 = pd.merge(df_datetesla,df_stocktesla,on="dateonly")
df_tesla2["average"]=(df_tesla2["High"]+df_tesla2["Low"])/2
df_tesla2
```

```
#Draw line graph with 2 y axes
fig,ax1=plt.subplots(figsize=(16,8))
ax2=plt.twinx(ax1)

ax1.plot(df_tesla2["dateonly"],df_tesla2['score'],label='sentiment score', color='orange',marker='^')
ax2.plot(df_tesla2["dateonly"],df_tesla2['High'],label='Highest stock price', color='blue', marker='^')

ax2.set_ylim(1000,1120)

plt.grid()
handler1, label1 = ax1.get_legend_handles_labels()
handler2, label2 = ax2.get_legend_handles_labels()
ax1.legend(handler1 + handler2, label1 + label2)
```
